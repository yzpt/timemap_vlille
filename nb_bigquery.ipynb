```python
from google.cloud import bigquery
import os
import pandas as pd
import datetime
```

```python
os.environ["GOOGLE_APPLICATION_CREDENTIALS"] = "key.json"

client = bigquery.Client()

project_id = "zapart-data-vlille"
dataset_id = "vlille_dataset"
```

```python
table_ref = client.get_table(f'{project_id}.{dataset_id}.stations')
schema = table_ref.schema

for field in schema:
    print(field.name, field.field_type)
```

```python
sql = f"""
SELECT *
FROM `{project_id}.{dataset_id}.stations`
"""

df_stations = client.query(sql).to_dataframe()
```

```python
df_stations.head()
```

```python
# add a column 'geo' [latitude, longitude]
df_stations['geo'] = df_stations.apply(lambda row: [row['latitude'], row['longitude']], axis=1)
df_stations.head()
```
